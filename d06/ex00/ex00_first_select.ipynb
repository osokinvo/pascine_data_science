<h2>Select<h2>

```python
import pandas as pd
import sqlite3
```

<h3>create a connection to the database<h3>

```python
db_con = sqlite3.connect('checking-logs.sqlite.sqlite_copy')
```

<h3>the schema of the table pageviews<h3>

```python
pd.io.sql.read_sql(sql='PRAGMA table_info(pageviews);', con=db_con)
```

<h3>get only the first 10 rows of the table pageviews<h3>

```python
pd.io.sql.read_sql(sql='SELECT * FROM pageviews LIMIT 10;', con=db_con)
```

<h3>the subtable<h3>

<h4>uid and datetime are used<h4>

```python
sql_query="""
SELECT 
    uid, datetime
FROM 
    pageviews
"""
pd.io.sql.read_sql(sql=sql_query, con=db_con, index_col='uid')
```

<h4>user data (user_*) is used and not admin data<h4>

```python
sql_query="""
SELECT 
    uid, datetime
FROM 
    pageviews
WHERE 
    uid LIKE 'user_%'
"""
pd.io.sql.read_sql(sql=sql_query, con=db_con, index_col='uid')
```

<h4>sorted by uid in ascending order<h4>

```python
sql_query="""
SELECT 
    uid, datetime
FROM 
    pageviews
WHERE 
    uid LIKE 'user_%'
ORDER BY 
    uid ASC;
"""
pd.io.sql.read_sql(sql=sql_query, con=db_con, index_col='uid')
```

<h4>the index column is datetime<h4>

```python
sql_query="""
SELECT 
    uid, datetime
FROM 
    pageviews
WHERE 
    uid LIKE 'user_%'
ORDER BY 
    uid ASC;
"""
pd.io.sql.read_sql(sql=sql_query, con=db_con, index_col='datetime')
```

<h4>the name of the dataframe is pageviews<h4>

```python
sql_query="""
SELECT 
    uid, datetime
FROM 
    pageviews
WHERE 
    uid LIKE 'user_%'
ORDER BY 
    uid ASC;
"""
pageviews = pd.io.sql.read_sql(sql=sql_query, con=db_con, index_col='datetime')
pageviews.name = 'pageviews'
pageviews
```

<h3>close the connection to the database<h3>

```python
db_con.close()
```
