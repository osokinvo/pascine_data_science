<h2>Preprocessing<h2>

<h3>read the CSV file and make ID the index column<h3>

```python
import pandas as pd
df = pd.read_csv('auto.csv', index_col=0)
df
```

<h3>count the number of observations using the method count()<h3>

```python
df.count()
```

<h3>drop the duplicates, taking into account only the following columns: CarNumber, Make_n_model, Fines<h3>

```python
df.drop_duplicates(subset=['CarNumber', 'Make_n_model', 'Fines'], keep='last', inplace=True)
df
```

```python
df.count()
```

<h3>work with missing values<h3>

<h4>check how many values are missing from each column<h4>

```python
df.isna().sum(axis=0)
```

<h4>drop all the columns with over 500 missing values<h4>

```python
df.dropna(axis=1, thresh=(df.shape[0] - 500), inplace=True)
df
```

```python
df.isna().sum(axis=0)
```

<h4>replace all the missing values in the Refund column with the previous value in that column<h4>

```python
df['Refund'].fillna(method='pad', inplace=True)
df
```

```python
df.isna().sum(axis=0)
```

<h4>replace all the missing values in the Fines column with the mean value of this column<h4>

```python
mean_fines = df.Fines.mean(skipna=True)
mean_fines
```

```python
df.Fines.fillna(value=mean_fines, inplace=True)
df
```

<h3>split and parse the make and model<h3>

```python
df[['Make', 'Model']] = pd.DataFrame(df['Make_n_model'].apply(str.split).tolist(), index=df.index)
df
```

```python
df.drop(labels='Make_n_model', axis=1, inplace=True)
df
```

```python
df.to_json('auto.json', orient='records')
```
