<h2>Selects and aggregations<h2>

<h3>load the JSON file that you created in the previous exercise into a dataframe<h3>

```python
import pandas as pd
df = pd.read_json('auto.json', orient='records')
df.set_index('CarNumber', inplace=True)
df
```

<h3>selects<h3>

<h4>the rows only where the fines are more than 2,100<h4>

```python
df.loc[df.Fines > 2100]
```

<h4>the rows only where the fines are more than 2,100 and the refund equals 2<h4>

```python
df.loc[(df.Fines > 2100) & (df.Refund == 2)]
```

<h4>the rows only where the models are from the list: [’Focus’, ’Corolla’]<h4>

```python
df.loc[df.Model.apply(lambda x: x in ['Focus', 'Corolla'])]
```

<h4>the rows only where the car number is from the list: [’Y7689C197RUS’, ’92928M178RUS’, ’7788KT197RUS’, ’H115YO163RUS’, ’X758HY197RUS’]<h4>

```python
car_num_list = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
df[df.index.isin(car_num_list)]
```

<h3>the aggregations with the make and the model<h3>

<h4>the median fines grouped by the make<h4>

```python
df.groupby('Make').Fines.agg('median')
```

<h4>the median fines grouped by the make and the model<h4>

```python
df.groupby(['Make','Model']).Fines.agg('median')
```

<h4>the number of fines grouped by the make and the model<h4>

```python
df.groupby(['Make','Model']).Fines.agg('count')
```

<h4>the minimum and the maximum fines grouped by the make and the model<h4>

```python
df.groupby(['Make','Model']).Fines.agg(['min', 'max'])
```

<h4>the standard deviation of the fines grouped by the make and the model<h4>

```python
df.groupby(['Make','Model']).Fines.agg('std')
```

<h3>the aggregations with the car number<h3>

<h4>the car numbers grouped by the number of the fines in descending order<h4>

```python
df.reset_index()[['CarNumber', 'Fines']].groupby('CarNumber', as_index=False).count().groupby(['Fines', 'CarNumber']).first().iloc[::-1]
```

<h4>all the rows corresponding to the top-1 car number<h4>

```python
df.loc[df.index[0]]
```

<h4>the car numbers grouped by the sum of the fines in descending order<h4>

```python
df.reset_index()[['CarNumber', 'Fines']].groupby(by=['CarNumber'], as_index=False).sum().groupby(['Fines', 'CarNumber']).first().iloc[::-1]
```
