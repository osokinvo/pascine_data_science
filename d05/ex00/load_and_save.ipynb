<h3>assign the following names to the column: datetime, user<h3>

```python
import pandas as pd

name_col_df = pd.read_csv('feed-views.log', sep="\t", names=['datetime','user'], engine='python')
name_col_df
```

<h3>Filter the rows with index 2 and 3 using the argument skiprows<h3>

```python
skiprows_df = pd.read_csv("feed-views.log", sep='\t', names=['datetime','user'], skiprows=[2,3], engine='python')
skiprows_df
```

<h3>filter the last 2 rows from the footer using the argument skipfooter<h3>

```python
skipfooter_auto_df = pd.read_csv("feed-views.log", sep='\t', names=['datetime','user'], skipfooter=2, engine='python')
skipfooter_auto_df
```
